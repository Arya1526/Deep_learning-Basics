```python
import sys
import pandas as pd

!{sys.executable} -m pip install langchain-text-splitters pandas

from langchain_text_splitters import RecursiveCharacterTextSplitter

text = """
Artificial Intelligence is transforming industries across the world. It helps automate repetitive tasks and improve efficiency in businesses. AI systems can analyze large amounts of data quickly and accurately.

Machine Learning is a branch of AI that enables systems to learn from data. Instead of being explicitly programmed, models improve their performance through experience. Popular algorithms include regression, decision trees, and neural networks.

Deep Learning is a subset of machine learning based on neural networks. It is widely used in image recognition, natural language processing, and speech recognition. Large datasets and powerful GPUs are often required.

Python is one of the most popular programming languages for AI and data science. Libraries such as NumPy, Pandas, Matplotlib, and Scikit-learn simplify development. Python syntax is also beginner friendly.

Data preprocessing is an important step in machine learning projects. Missing values, duplicate rows, and inconsistent formats must be handled carefully. Clean data improves model accuracy significantly.

Natural Language Processing allows computers to understand human language. Applications include chatbots, translation systems, and sentiment analysis. NLP models are improving rapidly with transformer architectures.

Computer Vision focuses on enabling machines to interpret visual data. Facial recognition, object detection, and autonomous vehicles use computer vision techniques. Convolutional neural networks are commonly applied.

Cloud computing provides scalable resources for AI applications. Services like AWS, Azure, and Google Cloud support model training and deployment. This reduces infrastructure costs for companies.

Cybersecurity is becoming more important as digital systems expand. AI can help detect fraud, malware, and unusual network behavior. However, attackers also use AI techniques for advanced threats.

The future of AI includes advancements in robotics, healthcare, education, and automation. Ethical considerations such as privacy, fairness, and transparency are essential. Responsible AI development is a major global focus.
"""

with open("sample.txt", "w") as file:
    file.write(text)

with open("sample.txt", "r") as file:
    content = file.read()

chunk_sizes = [200, 500, 1000]

for size in chunk_sizes:

    print(f"\n{'='*60}")
    print(f"Chunk Size: {size}")
    print(f"{'='*60}")

    splitter = RecursiveCharacterTextSplitter(
        chunk_size=size,
        chunk_overlap=40
    )

    chunks = splitter.split_text(content)

    print("Total Chunks:", len(chunks))

    chunk_data = []

    for i, chunk in enumerate(chunks):
        chunk_data.append({
            "Chunk Number": i + 1,
            "Character Count": len(chunk)
        })

    df = pd.DataFrame(chunk_data)

    print("\nChunk Character Counts:\n")
    print(df)

    print("\nSample Chunks:\n")

    for i in range(min(3, len(chunks))):
        print(f"Chunk {i+1}:\n")
        print(chunks[i])
        print("-" * 60)
```
