```python
import sys

!{sys.executable} -m pip install sentence-transformers langchain langchain-community scikit-learn

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity


documents = [
    "Virat Kohli scored a century in the cricket match.",
    "The bowler delivered a fast yorker to take the wicket.",
    "Cricket fans celebrated the thrilling last over finish.",
    "The team practiced batting and fielding before the tournament.",

    "Cooking pasta requires boiling water and preparing sauce.",
    "The chef added spices to improve the flavor of the curry.",
    "Baking cakes becomes easier with accurate measurements.",

    "Python is widely used for machine learning projects.",
    "Functions and loops are basic programming concepts.",
    "Debugging code helps programmers identify software errors."
]

queries = [
    "Cricket match highlights",
    "How to cook curry",
    "Programming in Python"
]


model = SentenceTransformer("all-MiniLM-L6-v2")

document_embeddings = model.encode(documents)


def manual_search(query, k=3):

    query_embedding = model.encode([query])

    similarity_scores = cosine_similarity(
        query_embedding,
        document_embeddings
    )[0]

    top_indices = similarity_scores.argsort()[::-1][:k]

    return [documents[i] for i in top_indices]


def retriever_search(query, k=3):

    query_embedding = model.encode([query])

    similarity_scores = cosine_similarity(
        query_embedding,
        document_embeddings
    )[0]

    top_indices = similarity_scores.argsort()[::-1][:k]

    return [documents[i] for i in top_indices]


for query in queries:

    print("\nQUERY:", query)

    manual_results = manual_search(query)

    retriever_results = retriever_search(query)

    print("\nManual Search Results:")
    for result in manual_results:
        print("-", result)

    print("\nRetriever Results:")
    for result in retriever_results:
        print("-", result)

    if manual_results == retriever_results:
        print("\nBoth results are the same")
    else:
        print("\nResults are different")
```
