```python


from langchain_core.prompts import PromptTemplate
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser
import os

os.environ["GROQ_API_KEY"] = "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"

prompt = PromptTemplate(
    input_variables=["role", "topic", "format"],
    template="""
You are a {role}.

Explain the topic: {topic}

Give the response in {format} format.
"""
)

model = ChatGroq(
    model_name="llama3-8b-8192",
    temperature=0.7
)

parser = StrOutputParser()

chain = prompt | model | parser

inputs = [
    {
        "role": "Teacher",
        "topic": "Python Functions",
        "format": "bullet points"
    },
    {
        "role": "Data Scientist",
        "topic": "Machine Learning",
        "format": "short paragraph"
    },
    {
        "role": "Software Engineer",
        "topic": "API Development",
        "format": "step-by-step"
    },
    {
        "role": "Cybersecurity Expert",
        "topic": "Phishing Attacks",
        "format": "numbered list"
    }
]

for i, data in enumerate(inputs, start=1):
    print(f"\nOutput {i}\n")

    try:
        result = chain.invoke(data)
        print(result)

    except Exception as e:
        print("Error:", e)
```
