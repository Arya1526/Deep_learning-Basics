```python
import sys

!{sys.executable} -m pip install langchain langchain-community langchain-text-splitters pypdf

from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

pdf_path = "/Users/Aryaanil/Documents/survey cdc.pdf"

loader = PyPDFLoader(pdf_path)
documents = loader.load()

print("Total Pages Loaded:", len(documents))

print("\nMetadata of Page 1:\n")
print(documents[0].metadata)

print("\nFirst 300 Characters of Page 1:\n")
print(documents[0].page_content[:300])

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=400,
    chunk_overlap=40
)

chunks = text_splitter.split_documents(documents)

print("\nTotal Chunks Created:", len(chunks))

print("\nShowing 3 Consecutive Chunks:\n")
for i in range(3):
    print(f"\nChunk {i+1}:\n")
    print(chunks[i].page_content)
    print("-" * 80)
```
