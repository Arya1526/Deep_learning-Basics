```python
import tensorflow as tf

tensor = tf.random.uniform((5, 4), minval=1, maxval=10)

print("Original Tensor:")
print(tensor)

mean_value = tf.reduce_mean(tensor)
print("\nMean:", mean_value.numpy())

max_value = tf.reduce_max(tensor)
print("Max:", max_value.numpy())

std_value = tf.math.reduce_std(tensor)
print("Standard Deviation:", std_value.numpy())

reshaped_tensor = tf.reshape(tensor, (4, 5))

print("\nReshaped Tensor (4, 5):")
print(reshaped_tensor)
```
