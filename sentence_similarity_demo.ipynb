```python
import sys
import pandas as pd

!{sys.executable} -m pip install sentence-transformers scikit-learn pandas

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

sentences = [
    "Virat Kohli scored a century in the cricket match.",
    "The bowler delivered a fast yorker to take the wicket.",
    "Cricket fans celebrated the thrilling last over finish.",
    "The team practiced batting and fielding before the tournament.",

    "Cooking pasta requires boiling water and preparing sauce.",
    "The chef added spices to improve the flavor of the curry.",
    "Baking cakes becomes easier with accurate measurements.",

    "Python is widely used for machine learning projects.",
    "Functions and loops are basic programming concepts.",
    "Debugging code helps programmers identify software errors."
]

topics = [
    "Cricket",
    "Cricket",
    "Cricket",
    "Cricket",
    "Cooking",
    "Cooking",
    "Cooking",
    "Programming",
    "Programming",
    "Programming"
]

model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(sentences)

similarity_matrix = cosine_similarity(embeddings)

rows = []

for i in range(len(sentences)):
    for j in range(i + 1, len(sentences)):
        rows.append({
            "Sentence 1": sentences[i],
            "Sentence 2": sentences[j],
            "Topic 1": topics[i],
            "Topic 2": topics[j],
            "Cosine Similarity": round(similarity_matrix[i][j], 3)
        })

df = pd.DataFrame(rows)

print("\nCosine Similarity Between Sentence Pairs:\n")
print(df.to_string(index=False))

same_topic = []
cross_topic = []

for i in range(len(sentences)):
    for j in range(i + 1, len(sentences)):

        score = similarity_matrix[i][j]

        if topics[i] == topics[j]:
            same_topic.append(score)
        else:
            cross_topic.append(score)

avg_same = sum(same_topic) / len(same_topic)
avg_cross = sum(cross_topic) / len(cross_topic)

print("\nAverage Similarity (Same Topic):", round(avg_same, 3))
print("Average Similarity (Cross Topic):", round(avg_cross, 3))

if avg_same > avg_cross:
    print("\nVerification Successful:")
    print("Same-topic sentences have higher similarity.")
else:
    print("\nVerification Failed:")
    print("Cross-topic similarity is unexpectedly higher.")
```
